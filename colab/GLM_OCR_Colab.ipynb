## 3. Carica un file (PDF o immagine)

```python
from google.colab import files
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
print(f"File caricato: {file_path}")
```

## 4. Esegui OCR (backend transformers_direct — veloce)

```python
from colab.glm_ocr_colab import ColabOCR

engine = ColabOCR(backend="transformers_direct")
markdown_text = engine.ocr(file_path)

print(markdown_text[:2000])  # anteprima
print(f"\n...\nTotale caratteri: {len(markdown_text)}")
```

## 6. Salva il risultato

```python
output_name = file_path.rsplit('.', 1)[0] + "_ocr.md"
with open(output_name, 'w', encoding='utf-8') as f:
    f.write(markdown_text)
files.download(output_name)
```
